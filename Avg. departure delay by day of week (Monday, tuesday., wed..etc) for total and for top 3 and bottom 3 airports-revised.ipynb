```python
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt

%matplotlib inline
flight_nyc=pd.read_csv('D:/Data Science Certification/attachment_Project_1_NYC-Flight_data/Project_1_NYC-Flight data/flight_data.csv')
flight_nyc_df=pd.DataFrame(flight_nyc)
#convert Time_hour column into a time stamp format 
flight_nyc_df['timestamp'] = pd.to_datetime(flight_nyc_df.time_hour, format='%d-%m-%Y %H:%M')
#created dictionary to map int value of weekday with actaul day of week 
dayOfWeek={0:'Monday', 1:'Tuesday', 2:'Wednesday', 3:'Thursday', 4:'Friday', 5:'Saturday', 6:'Sunday'}
flight_nyc_df['weekday']=flight_nyc_df['timestamp'].dt.dayofweek.map(dayOfWeek)
#flight_nyc_df.groupby('weekday').dep_delay.mean().plot(kind='bar')
#groupbyweekday_flight_nyc_df

flight_nyc_df.groupby('weekday').dep_delay.mean().plot()
groupbyweekday_flight_nyc=flight_nyc_df.groupby('weekday')

#Departure delay analysis for top 3 airports. We already indetified highest volume airports in terms of flight handling i.e. EWR, JFK, LGA.
#for EWR airport
filterd_flight_nyc_df=pd.DataFrame(flight_nyc[(flight_nyc.origin == 'EWR')])
filterd_flight_nyc_df_EWR=filterd_flight_nyc_df.groupby('weekday').dep_delay.mean()
filterd_flight_nyc_df_EWR_df=pd.DataFrame(filterd_flight_nyc_df_EWR)
filterd_flight_nyc_df_EWR_df=pd.DataFrame(filterd_flight_nyc_df_EWR).rename(columns={'dep_delay':'dep_delay_EWR'})
filterd_flight_nyc_df_EWR_df.plot()
#for JFK airport
filterd_flight_nyc_df=pd.DataFrame(flight_nyc[(flight_nyc.origin == 'JFK')])
filterd_flight_nyc_df_JFK=filterd_flight_nyc_df.groupby('weekday').dep_delay.mean()
filterd_flight_nyc_df_JFK_df=pd.DataFrame(filterd_flight_nyc_df_JFK)
filterd_flight_nyc_df_JFK_df=pd.DataFrame(filterd_flight_nyc_df_JFK).rename(columns={'dep_delay':'dep_delay_JFK'})
filterd_flight_nyc_df_JFK_df.plot()
#for LGA airport
filterd_flight_nyc_df=pd.DataFrame(flight_nyc[(flight_nyc.origin == 'LGA')])
filterd_flight_nyc_df_LGA=filterd_flight_nyc_df.groupby('weekday').dep_delay.mean()
filterd_flight_nyc_df_LGA_df=pd.DataFrame(filterd_flight_nyc_df_LGA).rename(columns={'dep_delay':'dep_delay_LGA'})
filterd_flight_nyc_df_LGA_df.plot()

#***************************BOTTOM THREE AIRPORT************************************

filterd_flight_nyc_df=pd.DataFrame(flight_nyc[(flight_nyc.dest == 'LEX')])
filterd_flight_nyc_df_LEX=filterd_flight_nyc_df.groupby('weekday').dep_delay.mean()
filterd_flight_nyc_df_LEX_df=pd.DataFrame(filterd_flight_nyc_df_LEX)
filterd_flight_nyc_df_LEX_df=pd.DataFrame(filterd_flight_nyc_df_LEX).rename(columns={'dep_delay':'dep_delay_LEX'})
filterd_flight_nyc_df_LEX_df.plot()

filterd_flight_nyc_df=pd.DataFrame(flight_nyc[(flight_nyc.dest == 'ANC')])
filterd_flight_nyc_df_ANC=filterd_flight_nyc_df.groupby('weekday').dep_delay.mean()
filterd_flight_nyc_df_ANC_df=pd.DataFrame(filterd_flight_nyc_df_ANC)
filterd_flight_nyc_df_ANC_df=pd.DataFrame(filterd_flight_nyc_df_ANC).rename(columns={'dep_delay':'dep_delay_ANC'})
filterd_flight_nyc_df_ANC_df.plot()

filterd_flight_nyc_df=pd.DataFrame(flight_nyc[(flight_nyc.dest == 'SBN')])
filterd_flight_nyc_df_SBN=filterd_flight_nyc_df.groupby('weekday').dep_delay.mean()
filterd_flight_nyc_df_SBN_df=pd.DataFrame(filterd_flight_nyc_df_SBN)
filterd_flight_nyc_df_SBN_df=pd.DataFrame(filterd_flight_nyc_df_SBN).rename(columns={'dep_delay':'dep_delay_SBN'})
filterd_flight_nyc_df_SBN_df.plot()



```
