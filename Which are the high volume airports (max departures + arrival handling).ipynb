```python
#Which are the high volume airports (max departures + arrival handling)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
flight_nyc=pd.read_csv('D:/Data Science Certification/attachment_Project_1_NYC-Flight_data/Project_1_NYC-Flight data/flight_data.csv')
flight_nyc_df=pd.DataFrame(flight_nyc)
#flight_nyc_df.dest.unique()
#flight_nyc_df.origin.unique()
origin_flight_count=flight_nyc.groupby('origin').flight.size()
origin_flight_count_df=pd.DataFrame(origin_flight_count).rename(columns={0: 'Total Flight at dest'})
dest_flight_count=flight_nyc.groupby('dest').flight.size()
total_flight_count=pd.concat([origin_flight_count_df,dest_flight_count], axis=1).rename(columns={0: 'Total Flight at origin'})
total_flight_count.fillna(0, axis=1)
total_flight_eachlocation=total_flight_count.sum(axis=1)
volume_flight_eachlocation_df=pd.DataFrame(total_flight_eachlocation).rename(columns={0:'total flight per location'}).sort(columns='total flight per location', ascending='true', axis=0)
volume_flight_eachlocation_df.tail(15).plot(kind='bar')

#total_flight_eachlocation_df.plot()












```
